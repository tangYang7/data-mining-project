```python
# fill the missing value
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values = 0, strategy='mean')
# col = ['glucose_concentration','blood_pressure', 'bmi']
col = ['glucose_concentration','blood_pressure', 'bmi', 'skin_fold_thickness','serum_insulin']
imp_mean.fit(data[col])
data[col] = imp_mean.transform(data[col])
print("The missing value is filled.")
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
```

```python
x = data.drop(['diabetes','p_id'], axis=1)
y = data['diabetes']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y)
x
```

```python
# RandomForestClassifier
print("RandomForestClassifier(): ")
rfc = RandomForestClassifier(max_depth=5, min_samples_split=3)
#Fit model on the training Data
rfc.fit(x_train,y_train)
#Make prediction
predictions = rfc.predict(x_test)
acc = accuracy_score(y_test,predictions)
print('Accuracy:{:.4%}'.format(acc))
f1 = f1_score(y_test,predictions)
print('F1_score:{:.4%}'.format(f1))

# MLPClassifier()
print("MLPClassifier(): ")
rfc = MLPClassifier(hidden_layer_sizes=(20,20),max_iter=200)
#Fit model on the training Data
rfc.fit(x_train,y_train)
#Make prediction
predictions = rfc.predict(x_test)
acc = accuracy_score(y_test,predictions)
print('Accuracy:{:.4%}'.format(acc))
f1 = f1_score(y_test,predictions)
print('F1_score:{:.4%}'.format(f1))
```

```python
# AdaBoostClassifier()
print("AdaBoostClassifier():")
rfc = AdaBoostClassifier()
#Fit model on the training Data
rfc.fit(x_train,y_train)
#Make prediction
predictions = rfc.predict(x_test)
acc = accuracy_score(y_test,predictions)
print('Accuracy:{:.4%}'.format(acc))
f1 = f1_score(y_test,predictions)
print('F1_score:{:.4%}'.format(f1))

# DecisionTreeClassifier
print("DecisionTreeClassifier():")
rfc = DecisionTreeClassifier()
#Fit model on the training Data
rfc.fit(x_train,y_train)
#Make prediction
predictions = rfc.predict(x_test)
acc = accuracy_score(y_test,predictions)
print('Accuracy:{:.4%}'.format(acc))
f1 = f1_score(y_test,predictions)
print('F1_score:{:.4%}'.format(f1))
```

```python
# GaussianNB()
print("GaussianNB():")
rfc = GaussianNB()
#Fit model on the training Data
rfc.fit(x_train,y_train)
#Make prediction
predictions = rfc.predict(x_test)
acc = accuracy_score(y_test,predictions)
print('Accuracy:{:.4%}'.format(acc))
f1 = f1_score(y_test,predictions)
print('F1_score:{:.4%}'.format(f1))
```
